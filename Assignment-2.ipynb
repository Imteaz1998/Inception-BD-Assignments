1. Email Validator

```python
email=str(input("Please insert an email address: "))
symbol=False

for i in email:
    if i =="@":
        symbol=True

if symbol and (email.endswith('.com') or email.endswith(".org")):
    print("Valid email")
else:
    print("Invalid Email")
```

2. Count word frequency in customer reviews

```python
review=str(input("Please enter a review: "))
review=review.lower()

splitted_words=review.split()
freq={}

for i in splitted_words:
    if i in freq:
        freq[i]=freq[i]+1
    else:
        freq[i]=1

print(freq)
```

3. Password Strength checker

```python
password=str(input("Enter a password: "))

has_upper=False
has_lower=False
has_digit=False
has_special=False

for i in password:
    if i.isupper():
        has_upper=True
    elif i.islower():
        has_lower=True
    elif i.isdigit():
        has_digit=True
    else:
        has_special=True

if has_upper and has_lower and has_digit and has_special:
    print("Strong password")
else:
    print("Weak password")
```

4. Username Generator

```python
import random
name=str(input("Enter a name: "))
name = name.replace(" ", "")
first_name = ""
last_name = ""

for i in range(len(name)):
    if i < 3:
        first_name += name[i].lower()
    if i >= len(name) - 3:
        last_name += name[i].lower()

random_number = str(random.randint(10, 99))
username = first_name + last_name + random_number
print(username)


    
 
```

5. SMS Word counter

```python
sms=str(input("Please enter a sms: "))
words=list(sms.split())

words_count=len(words)
char_count=len(sms)

if char_count>=160:
    print(f"Words={words_count}, Characters={char_count}, Status = Exceed SMS limt")
else:
    print(f"Words={words_count}, Characters={char_count}, Status = Within limt")

    
    
```

6. Remove duplicates from contact list

```python
name_list=str(input("Please enter a sentence: "))
name_list=name_list.replace(" ","")
splitted_name=name_list.split(",")

freq={}

for i in splitted_name:
    if i in freq:
        freq[i]=freq[i]+1
    else:
        freq[i]=1

output = "" + ", ".join(freq.keys()) + ""
print(output)
    
```

7. Detect Suspicious words in Text

```python
t=str(input("Please enter a sentence: "))
splited_t=list(t.split())
banned_words=['spam', 'scam', 'fake']

for word in splited_t:
    if word.lower() in banned_words:
        print("Suspicous content detected")
        break
```

9. Count hashtags and mentions

```python
post=str(input("Please enter a social media post: "))
splitted_post=list(post)
freq_at=0
freq_hash=0

for i in splitted_post:
    if i == "@":
        freq_at=freq_at+1
    elif i == "#":
        freq_hash=freq_hash+1
    

print(f"Hashtags ={freq_hash}, Mentions = {freq_at}")
```

10. Word Reversal for data encryption

```python
input_text = str(input("Please enter a sentence: "))
splitted_text = input_text.split()

result = []

for i in splitted_text:
    result.append(i[::-1])

output = " ".join(result)
print(output)
```
