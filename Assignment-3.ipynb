# 1. Inventory Price Calculator

```python
price=input("Please enter the prices of the items with comma: ")
price_list=price.split(',')

price_list = [int(i) for i in price_list]
sum=0
for i in price_list:
    sum=sum+i
avg=sum/len(price_list)
output_list=[sum,avg]
output=tuple(output_list)
print(output)
```

# 2. Unique Visitor Tracker

```python

input_list=input("Please enter the visitor tracker with comma: ")


def count_unique_visitors(visitor_list):
    
    """
    This function checks unique visitor
    
    """
    visitor_list = input_list.split(',')
    list_d=[]
    for i in visitor_list:
      if i not in list_d:
        list_d.append(i)
    print(f"{len(list_d)} (The unique ids are {','.join(list_d)})")


count_unique_visitors(input_list)

    
        

```

# 3. Product Category Counter

```python
products_name=input("Please enter a review: ")


def count_categories(products):
    
    """
    This function checks Product Category Counter
    
    """
    products=products.lower()
    splitted_product=products.split(",")
    
    freq={}

    for i in splitted_product:
      item = i.strip()      # Remove spaces
      if item in freq:
            freq[item] += 1
      else:
            freq[item] = 1

    print(freq)


count_categories(products_name)
```

# 4. Temprature Analytics

```python
temprature=input("Please enter a temprature: ")

def analyze_tempratures(temp_list):

    """
    This function analyze temprature
    
    """
    
    splitted_temps=temp_list.split(",")
    
    high_list=[]
    normal_list=[]

    for i in splitted_temps:
        value=float(i.strip())
        if value>38:
            high_list.append(value)
        else:
            normal_list.append(value)
    print(f"High list(>38째C): {high_list}")
    print(f"Normal list (>38째C): {normal_list}")
    full_output_tuple = (high_list, normal_list)
    print("Final output as tuple:", full_output_tuple)
    
analyze_tempratures(temprature)
```

# 5. Menu price lookup

```python
# Input values= {"Coffee":3.50,"Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}
user_input = input("Enter the menu dictionary: ")
item_name = input("Enter item name: ")

def get_price(menu, item):

    """
    This function generates price
    
    """
    if item in menu:
        return menu[item]
    else:
        return "Item not found"

menu = {}


cleaned = user_input.strip()[1:-1]

pairs = cleaned.split(",")

for p in pairs:
    key, value = p.split(":")
    key = key.strip().strip('"').strip("'")    
    
    value = float(value.strip())                
    
    menu[key] = value

print(get_price(menu, item_name))

```

# 8. Hashtag Generator

```python
input_string=input("Please Enter a string: ")


def generate_hashtag(sentence):
    
    """
    This function generates hashtags
    
    """
    x=sentence.split()
    y=[]
    for i in x:
      y.append('#'+i)
    print(y)
generate_hashtag(input_string)
   
```
