# Assignment 1 Temprature Check

```python
temprature=int(input("Please enter temprature in centrigrade scale: "))

if temprature <10:
    print("It's too cold!")
elif temprature >= 10 and temprature<=25:
    print("It's cool outside")
else:
    print("It's hot outside!")
```

# Assignment 2 ATM Withdrawal System

```python
balance=int(input("Please enter the balance of your account: "))
amount=int(input("Please enter the amount you wish to withdraw: "))

if amount>balance:
    print("Insufficient Balance" )
else:
    print("Transaction Successful")
```

# Assignment 3 Student Grade Calculator

```python
marks=int(input("Please enter your marks: "))

if marks>=90:
    print("A+")
elif 80<=marks<=89:
    print('A')
elif 70<=marks<=79:
    print('B')
elif 60<=marks<=69:
    print('C')
else:
    print('Fail')
```

# Assignment 4: Bus Ticket Discount

```python
age=int(input("Please Enter your age: "))

if age<5:
    print("Free")
elif 5<=age<=18:
    print("50% Discount")
elif age>=60:
    print("30% Discount")
else:
    print("Full price")
```

# Assignment 5: Resturant Bill with Discount

```python
bill_amount=int(input("Please Enter the amount: "))

if bill_amount > 1000:
    finale_bill=bill_amount-bill_amount*0.1
    print("Your bill is: ", finale_bill)
else:
    print("Your bill is: ", bill_amount)
```

# Assignment 7: Traffic Light System

```python
color=input("Please enter a color name: ")

if color == "red":
    print("Stop")
elif color == "yellow":
    print("ready to go")
elif color == "green":
    print("GO")
else:
    print("Invalid color")
```

# Assignment 8: Leap year checker

```python
year= int(input("Please enter a year: "))

if year%400==0:
    print("Leap year")
elif year%4==0 and year%100!=0:
    print("Leap year")
else:
    print("Not a Leap year")
```

# Assignment 10: Electricity Bill Calculator

```python
units=int(input("Enter please the consumed units: "))

if units<=100:
    bill=units*5
    print("Your bill is BDT ",bill," Taka")
elif 101<=units<=200:
    bill=units*7
    print("Your bill is BDT ",bill," Taka")
else:
    bill=units*10
    print("Your bill is BDT ",bill," Taka")

```
